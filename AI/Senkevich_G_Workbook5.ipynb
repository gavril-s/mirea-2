```python
import math

class Trig:
    def __init__(self, deg):
        self.deg = deg
    
    
    def degtorad(self):
        return self.deg * math.pi / 180
    
    def sin(self):
        return round(math.sin(self.degtorad()), 2)
    
    def cos(self):
        return round(math.cos(self.degtorad()), 2)
    
    def tg(self):
        return round((math.sin(self.degtorad())/math.cos(self.degtorad())), 2)
    
    def arcsin(self):
        return round(math.asin(self.degtorad()), 2)
    
    def arccos(self):
        return round(math.acos(self.degtorad()), 2)
    
    def arctg(self):
        return round(math.atan(self.degtorad()), 2)
    
d = Trig(30)
print(d.sin())
print(d.cos())
print(d.tg())
print(d.arcsin())
print(d.arccos())
print(d.arctg())
```

```python
class Tree:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data
        
    def Insert(self, data):
        if (data < self.data):
            if self.left == None:
                self.left = Tree(data)
            else:
                self.left.Insert(data)
        else:
            if self.right == None:
                self.right = Tree(data)
            else:
                self.right.Insert(data)
                
    def PreOrderPrint(self):
        print(self.data)
        if self.left != None:
            self.left.PreOrderPrint()
        if self.right != None:
            self.right.PreOrderPrint()
            
t = Tree(1)
t.Insert(0)
t.Insert(3)
t.PreOrderPrint()
```

```python
class Tree:
    def __init__(self, data, left, right):
        self.data = data
        self.left = left
        self.right = right
t = Tree("a", Tree("b", "d", "e"), Tree("c", "f", None))
print("Корень - ", t.data)
print("Левое поддерево:\n", t.left.data)
print(t.left.left, " ", t.left.right)
print("Правое поддерево:\n", t.right.data)
print(t.right.left, " ", t.right.right)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import tree, metrics
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split

url = r'https://raw.githubusercontent.com/aniruddhachoudhury/Red-Wine-Quality/master/winequality-red.csv'
dataset = pd.read_csv(url)

print(dataset.head())

plt.scatter(dataset['fixed acidity'], dataset['volatile acidity'], color='b')
plt.xlabel('fixed acidity')
plt.ylabel('volatile acidity')
plt.show()

x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state = 0)
regressor = DecisionTreeRegressor()
regressor.fit(x_train, y_train)
tree.plot_tree(regressor)
y_pred = regressor.predict(x_test)

df = pd.DataFrame({'Actual':y_test, 'Predicted':y_pred})
print(df)
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
metrics.mean_absolute_error(y_test, y_pred) / np.average(y) * 100
```
