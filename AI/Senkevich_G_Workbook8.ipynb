```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

k = 3
X = np.array([
    [5, 3],
    [10, 15],
    [15, 12],
    [24, 10],
    [30, 45],
    [85, 70],
    [71, 80],
    [60, 78],
    [55, 52],
    [80, 91]
])

kmeans = KMeans(n_clusters = k)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
centers = kmeans.cluster_centers_

plt.axes().set_facecolor('0.9')
plt.scatter(X[:, 0], X[:, 1], c = y_kmeans, s = 50)
plt.scatter(centers[:, 0], centers[:, 1], c = 'white', s = 200, alpha = 0.9)
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

k = 3

iris = sns.load_dataset('iris')
x, y = iris.iloc[:, :-1], iris.iloc[:, -1]

plt.figure(figsize=(10, 7))
sns.scatterplot(x='petal_width', y='petal_length', data=iris, hue='species', s=70)
plt.legend(loc=2)
plt.grid()
plt.show()

kmeans = KMeans(n_clusters = k)
kmeans.fit(x)
y_kmeans = kmeans.predict(x)
centers = kmeans.cluster_centers_

plt.figure(figsize=(10, 7))
sns.scatterplot(x='petal_width', y='petal_length', data=iris, hue=y_kmeans, s=70)
plt.scatter(centers[:, 3], centers[:, 2], c = 'blue', s = 200, alpha = 0.9)
plt.legend(loc=2)
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import scipy.cluster.hierarchy as shc
from sklearn.cluster import AgglomerativeClustering

k = 3

iris = sns.load_dataset('iris')
x, y = iris.iloc[:, 2:-1], iris.iloc[:, -1]

plt.figure(figsize=(20, 20))
dend = shc.dendrogram(shc.linkage(x, method='ward'))
plt.show()

cluster = AgglomerativeClustering(n_clusters=k, affinity='euclidean', linkage='ward')
cluster.fit_predict(x)

plt.figure(figsize=(10, 7))
sns.scatterplot(x='petal_width', y='petal_length', data=iris, hue=cluster.labels_, s=70)
plt.legend(loc=2)
plt.grid()
plt.show()
```
