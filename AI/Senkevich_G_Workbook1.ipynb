```python
#4.3.2
from math import log, sqrt, sin, cos, e
from matplotlib import pyplot as plt

def f(x):
    return sqrt(1 + e**sqrt(x) + cos(x**2)) / abs(1 - sin(x)**3) + log(abs(2*x))

x = [i for i in range(1, 11)]
y = [f(i) for i in x]

plt.plot(x, y)
#plt.show()

plt.scatter(x[:len(x) // 2], y[:len(y) // 2])
```

```python
#4.3.3
import numpy as np
from math import cos, log, e
from matplotlib import pyplot as plt
from scipy.integrate import simps
from numpy import trapz

def f(x):
    return abs(cos(x * e**(cos(x) + log(x + 1))))
                   
x = np.arange(0, 10, 1)
y = np.array(list(map(f, x)))
                   
plt.grid()
plt.plot(x, y, c='r')
plt.fill_between(x, y)

area = trapz(y)
print('area:', area)
```

```python
#4.3.4
import numpy as np
from matplotlib import pyplot as plt
from datetime import datetime, date

AAPL = np.genfromtxt('data/AAPL_2021.csv', delimiter=',', dtype=None, encoding='UTF-8')[1:]
GOOG = np.genfromtxt('data/GOOG_2021.csv', delimiter=',', dtype=None, encoding='UTF-8')[1:]
MSFT = np.genfromtxt('data/MSFT_2021.csv', delimiter=',', dtype=None, encoding='UTF-8')[1:]

AAPL_dates = [datetime.strptime(row[0], '%Y-%m-%d') for row in AAPL]
AAPL_open = [float(row[1]) for row in AAPL]

GOOG_dates = [datetime.strptime(row[0], '%Y-%m-%d') for row in GOOG]
GOOG_open = [float(row[1]) for row in GOOG]

MSFT_dates = [datetime.strptime(row[0], '%Y-%m-%d') for row in MSFT]
MSFT_open = [float(row[1]) for row in MSFT]

plt.plot(AAPL_dates, AAPL_open, color='r', label='AAPL (Apple)')
plt.plot(GOOG_dates, GOOG_open, color='g', label="GOOG (Google)")
plt.plot(MSFT_dates, MSFT_open, color='b', label="MSFT (Microsoft)")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.title("AAPL, GOOG and MSFT prices in 2021")
plt.axis([date(2021, 1, 1), date(2022, 1, 1), 0, 350])
plt.grid()
plt.legend()
plt.show()
```

```python
#4.3.5
import re
import math

class Operation:
    def __init__(self, regex, function):
        self.regex = regex
        self.pattern = re.compile(regex)
        self.function = function
        
    def match(self, string):
        return bool(re.fullmatch(self.pattern, string))
    
    def eval(self, string):
        return self.function(string)

number_regex = "(\d+(\.\d+)?)"
    
def process_add(string):
    arg1, arg2 = map(float, string.split("+"))
    return arg1 + arg2

def process_sub(string):
    arg1, arg2 = map(float, string.split("-"))
    return arg1 - arg2

def process_mul(string):
    arg1, arg2 = map(float, string.split("*"))
    return arg1 * arg2

def process_div(string):
    arg1, arg2 = map(float, string.split("/"))
    return arg1 / arg2

def process_pow(string):
    arg1, arg2 = map(float, string.split("^"))
    return arg1 / arg2

def process_exp(string):
    arg1, arg2 = map(float, string[3:-1].split("+"))
    return math.e ** (arg1 + arg2)
                      
def process_sin(string):
    arg1, arg2 = map(float, string[4:-1].split("+"))
    return math.sin(arg1 + arg2)

def process_cos(string):
    arg1, arg2 = map(float, string[4:-1].split("+"))
    return math.cos(arg1 + arg2)

def process_none(string):
    return "Bad input"

add = Operation(
    number_regex + "\+" + number_regex,
    process_add
)
subtract = Operation(
    number_regex + "\-" + number_regex,
    process_sub
)
multiply = Operation(
    number_regex + "\*" + number_regex,
    process_mul
)
divide = Operation(
    number_regex + "\/" + number_regex,
    process_div
)
pow = Operation(
    number_regex + "\^" + number_regex,
    process_pow
) 
exp = Operation(
    "e\^\(" + number_regex + "\+" + number_regex + "\)",
    process_exp
) 
sin = Operation(
    "sin\("+ number_regex + "\+" + number_regex + "\)",
    process_sin
) 
cos = Operation(
    "cos\("+ number_regex + "\+" + number_regex + "\)",
    process_cos
) 
none = Operation("(.*?)", process_none)

operations = [add, subtract, multiply, divide, pow, exp, sin, cos, none]

inp = input().replace(" ", "")
for operation in operations:
    if operation.match(inp):
        print(operation.eval(inp))
        break
```
