```python
#2.3.1
import pandas as pd

def distance(a, b):
    return ((a["x"] - b["x"])**2 +
            (a["y"] - b["y"])**2 +
            (a["z"] - b["z"])**2)**0.5

p1_coords = [1, 2, 3]
p2_coords = [3, 2, 1]

p1 = pd.Series(data=p1_coords, index=["x", "y", "z"])
p2 = pd.Series(data=p2_coords, index=["x", "y", "z"])
distance(p1, p2)
```
